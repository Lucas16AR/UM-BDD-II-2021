# Visualización con Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Line Plot

```python
data = [1,2,3, 5,10, 2 ,20]
plt.plot(data)
plt.show()
```

```python
x = [100,200,300,400,500,1000,1100]
y = [1,2,3, 5,10, 2 ,20]
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Título")
plt.plot(x,y)
plt.show()
```

```python
x = ["A","B","C","D","E","F","G"]
y = [1,2,3, 5,10, 2 ,20]
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Título")
plt.plot(x,y)
plt.show()
```

## Scatter Plot

```python
x = [1,2,3, 5,10,20]
y = [1,5,8, 5,2,3]
plt.scatter(x,y,color='red') 
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Título")
```

```python
ax = [1,2,3, 5,10,20]
ay = [1,5,8, 5,2,3]
bx = [1,2,3]
by = [5,10,12]

plt.scatter(ax,ay,color='red',label='Valores A') 
plt.scatter(bx,by,color='blue',label='Valores A') 
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Título")
plt.legend(loc='best') 
plt.show()
```

## Bar plot

```python
x = [0,1,2,3,4]
y = [10,80,25,35,27]
plt.bar(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Título")
plt.show();
```

```python
x = ["A","B","C","D","E"]
y = [10,80,25,35,27]
plt.bar(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Título")
plt.show();
```

```python
x2 = [0.5,1.5,2.5,3.5,4.5]
y2 = [12,75,30,40,30]
plt.bar(x, y ,label="A",color="c",width=.5)
plt.bar(x2, y2, label="B", color="r",width=.5)
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Título")
plt.show()
```

## Pie chart

```python
labels = 'A', 'B', 'C', 'D'
sizes = [50, 10, 25, 15]
explode = (0, 0.1, 0.2, 0)
fig1, a = plt.subplots()
a.pie(sizes, explode=explode, labels=labels)
a.axis('equal')
plt.show()
```

```python
labels = 'A', 'B', 'C', 'D'
sizes = [50, 10, 25, 15]
explode = (0, 0.5, 0.3, 0)
fig1, a = plt.subplots()
a.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',)
a.axis('equal')
plt.show()
```

## Histrogram

```python
x = [1,2,3, 5,10,20,2,1,2,19,10,10,4,3,10]
plt.hist(x,rwidth=0.9) 
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Título") 
plt.show();
```
